```python
# Import dependencies
import pandas as pd
import gmaps
import matplotlib.pyplot as plt

# Import requests library
import requests

# Import api key
from config import g_key

# Import datetime
from datetime import datetime
```

```python
# Import weatherpy_challenge.csv
file_to_load = 'weather_data/WeatherPy_Challenge.csv'

vacation_df = pd.read_csv(file_to_load)

vacation_df.head()
```

```python
# Configure gmaps to use your Google API key
gmaps.configure(api_key = g_key)
```

```python
# Dropping nan rows 
vacation_df = vacation_df.dropna()
```

```python
# Confirm nan rows were dropped
vacation_df.notnull().sum()
```

```python
# Prompt cutomer for temperature and weather preferences
min_temp = float(input('What is the minimum temperature you would like for your trip? '))
max_temp = float(input('What is the maximum temperature you would like for your trip? '))
rain = input('Do you want it to be raining? (yes/no) ')
snow = input('Do you want it to be snowing? (yes/no) ')
```

```python
# Get preferred data from temperature preferences
preferred_cities_df = vacation_df.loc[(vacation_df['Max Temp'] <= max_temp) & (vacation_df['Max Temp'] > min_temp)]

preferred_cities_df.head()
```

```python
# Get data based on answers about weather preferences

if ((rain == 'yes') and (snow == 'no')):
    rain_cities_df = preferred_cities_df.loc[preferred_cities_df['Rain Inches (last 3 hrs)'] > 0.0]
    df = rain_cities_df

elif ((rain == 'no') and (snow == 'yes')):
    snow_cities_df = preferred_cities_df.loc[preferred_cities_df['Snow Inches (last 3 hrs)'] > 0.0]
    df = snow_cities_df

elif ((rain == 'yes') and (snow == 'yes')):
    rain_snow_cities_df = preferred_cities_df.loc[(preferred_cities_df['Rain Inches (last 3 hrs)'] > 0.0) & (preferred_cities_df['Snow Inches (last 3 hrs)'] > 0.0)]
    df = rain_snow_cities_df

elif ((rain == 'no') and (snow == 'no')):
    no_rain_snow_cities_df = preferred_cities_df.loc[(preferred_cities_df['Rain Inches (last 3 hrs)'] == 0.0) & (preferred_cities_df['Snow Inches (last 3 hrs)'] == 0.0)]
    df = no_rain_snow_cities_df

else:
    print('Please confirm you answered all questions correctly.')

if (df.empty is True):
    print('No cities match your criteria.')

else:
    print(df.head())
```

```python
# Create DataFrame called hotel_df to store hotel names along with city, country, max temp, and coordinates
hotel_df = df[['City', 'Country', 'Max Temp', 'Current Weather', 'Lat', 'Lng']].copy()

# Add a new column for hotel name
hotel_df['Hotel Name'] = ''

hotel_df.head()
```

```python
# Set parameters to search for a hotel
params = {
    'radius': 5000,
    'type': 'lodging',
    'key': g_key
}
```

```python
# Iterate through the DataFrame
for index, row in hotel_df.iterrows():
    
    # Get the latitude and longitude
    lat = row['Lat']
    lng = row['Lng']
    
    # Add the lat and lon to location key for the params dictionary
    params['location'] = f'{lat}, {lng}'
    
    # Use the search term: lodging and our lat and lon
    base_url = 'https://maps.googleapis.com/maps/api/place/nearbysearch/json'
    
    # Make request and get the JSON data from the search
    hotels = requests.get(base_url, params = params).json()
    
    # Grab the first hotel from the results and store the name
    try:
        hotel_df.loc[index,'Hotel Name'] = hotels['results'][0]['name']
        
    except (IndexError):
        print('Hotel not found... skipping.')
```

```python
# Call rows of hotel_df
hotel_df.head()
```

```python
# Add info_box_template
info_box_template = """
<dl>
<dt>Hotel Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
<dt>Current Weather</dt><dd>{Current Weather} and {Max Temp} Â°F</dd>
</dl>
"""
```

```python
# Store the DataFrame Row
hotel_info = [info_box_template.format(**row) for index, row in hotel_df.iterrows()]
```

```python
# Add a heatmap of temperature for the vacation spots
# Get the lat, lon, and max temp
locations = hotel_df[['Lat', 'Lng']]
max_temp = hotel_df['Max Temp']

# Assign the figure variable
fig = gmaps.figure(center = (30.0, 31.0), zoom_level = 1.5)

# Assign the heatmap variable
heat_layer = gmaps.heatmap_layer(locations, weights = max_temp, dissipating = False, 
                                max_intensity = 300, point_radius = 4)

# Assign the marker variable
marker_layer = gmaps.marker_layer(locations, info_box_content = hotel_info)

# Add the heatmap and marker layer
fig.add_layer(heat_layer)
fig.add_layer(marker_layer)

# Call the figure to plot the data
fig
```

```python
# Create the output csv file
output_data_file = 'weather_data/WeatherPy_vacation.csv'

hotel_df.to_csv(output_data_file, index_label = 'City_ID')
```
